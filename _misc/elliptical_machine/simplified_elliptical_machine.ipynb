```python
x, y, r, a, theta = var('x y r a theta', domain='real')
```

```python
solutions = solve([a^2 == r^2 + x^2 - 2*r*x*cos(theta)], x)
```

```python
solutions[0]
```

```python
solutions[1]
```

```python
A = 2
R = 1
```

```python
plot(x.subs(solutions[0]).subs(a = A, r = R), (theta,0,2*pi)) + plot(-R-A*(1-cos(theta))/2, (theta,0,2*pi),color="red")
```

```python
plot(x.subs(solutions[1]).subs(a = 2, r = 1), (theta,0,2*pi)) + plot(R+A*(1+cos(theta))/2, (theta,0,2*pi),color="red")
```

```python
s = (r*exp(I*theta) + x)/2
```

```python
s1 = s.subs(solutions[0])
```

```python
s2 = s.subs(solutions[1])
```
