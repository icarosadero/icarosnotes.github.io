```python
x,y,r,a,C,b,theta,gamma = var('x y r a C b theta gamma', domain="real")
assume(C>r)
assume(b>a)
```

```python
var('x y r a C b theta gamma')
eq1 = (x - r*cos(theta))^2 + (y - r*sin(theta))^2 == a^2
eq2 = (x - C*cos(gamma))^2 + (y - C*sin(gamma))^2 == b^2
solutions = solve([eq1, eq2], x, y)
```

```python
len(solutions)
```

```python
X = solutions[0][0].simplify_full()
Y = solutions[1][0].simplify_full()
```

```python
show(X)
```

```python
show(Y)
```

```python
S = x^2 + y^2
show(S.subs(solutions[0]).simplify_full())
```

```python
z = x + I*y
z1 = z.subs(solutions[0]).full_simplify()
z2 = z.subs(solutions[1]).full_simplify()
```

```python
show(z2)
```

```python
s1 = (z1 + r*cos(theta) + I*r*sin(theta))/2
s2 = (z2 + r*cos(theta) + I*r*sin(theta))/2
```
