# Symbolc Solution

```python
x,y,r,a,C,b,theta,gamma = var('x y r a C b theta gamma', domain="real")
assume(C>r)
assume(b>a)
assume(a>0)
assume(r>0)
```

```python
var('x y r a C b theta gamma')
eq1 = (x - r*cos(theta))^2 + (y - r*sin(theta))^2 == a^2
eq2 = (x - C*cos(gamma))^2 + (y - C*sin(gamma))^2 == b^2
```

```python
solutions = solve([eq1, eq2], x, y)
```

```python
for pair in solutions:
    for v in pair:
        print((v.simplify_full()))
```

```python
z = x + I*y
z1 = z.subs(solutions[0]).full_simplify()
z2 = z.subs(solutions[1]).full_simplify()
```

```python
s1 = (z1 + r*cos(theta) + I*r*sin(theta))/2
s2 = (z2 + r*cos(theta) + I*r*sin(theta))/2
```

# Asymptotic Analysis

```python
#Taking the limit of the solutions as b-> infinity
H,Q = var("H Q", domain="real")
s1_limit = limit(s1.subs(cos(gamma), Q).subs(sin(gamma), H), H = infinity)
s2_limit = limit(s1.subs(cos(gamma), Q).subs(sin(gamma), H), H = infinity)
```
