# Symbolc Solution

```python
x,y,r,a,C,b,theta,gamma,H,Q = var('x y r a C b theta gamma H Q', domain="real")
assume(C>r)
assume(b>a)
assume(a>=r)
assume(r>0)
```

```python
var('x y r a C b theta gamma')
eq1 = (x - r*cos(theta))^2 + (y - r*sin(theta))^2 == a^2
eq2 = (x - Q)^2 + (y - H)^2 == b^2
```

```python
solutions = solve([eq1, eq2], x, y)
```

```python
for pair in solutions:
    for v in pair:
        print((v.simplify_full()))
```

```python
z = x + I*y
z1 = z.subs(solutions[0]).full_simplify()
z2 = z.subs(solutions[1]).full_simplify()
```

```python
s1 = (z1 + r*cos(theta) + I*r*sin(theta))/2
s2 = (z2 + r*cos(theta) + I*r*sin(theta))/2
```

# Asymptotic Analysis

```python
#Taking the limit of the solutions as b-> infinity
y_0 = var("y_0", domain="real")
assume(a>r)
s1_limit = limit(s1.subs(b=H-y_0).full_simplify(), H=infinity).full_simplify()
s2_limit = limit(s2.subs(b=H-y_0).full_simplify(), H=infinity).full_simplify()
```

```python
latex(s1_limit)
```

```python
latex(s2_limit)
```

```python
s1_limit.imag_part()
```

```python
eq3 = (y^2-(1/2)*(r*sin(theta)+y_0))^2==(r*sin(theta)-y_0)^2 - a^2
eq3_xy_0 = eq3.full_simplify().subs({sin(theta): sqrt(1-(x/r)^2)}).subs(y_0=0).full_simplify()
```

```python
latex(eq3_xy_0)
```

```python
A = 2
R = 1
P = A/2
S = R/2

s2_limit_plot = s2_limit.subs(
    y_0 = 0,
    r = R,
    a = A
)

legend_label = f"$y_0 = 0; r = {R}; a = {A}$"

plot = parametric_plot((s2_limit_plot.real_part(),s2_limit_plot.imag_part()),(theta,0,2*pi), legend_label=legend_label)

plot_2 = implicit_plot((x/P-1)^2 + (y/S)^2 == 1, (x,0,A), (y,-S,S), color="red", legend_label="ellipse")
```
