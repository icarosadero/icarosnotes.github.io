```python
x, y, r, a, theta = var('x y r a theta', domain='real')
```

```python
solutions = solve([a^2 == r^2 + x^2 - 2*r*x*cos(theta)], x)
```

```python
show(solutions[0])
```

```python
show(solutions[1])
```

```python
A = 2
R = 1
```

```python
plot(x.subs(solutions[0]).subs(a = A, r = R), (theta,0,2*pi)) + plot(-R-A*(1-cos(theta))/2, (theta,0,2*pi),color="red")
```

```python
plot(x.subs(solutions[1]).subs(a = 2, r = 1), (theta,0,2*pi)) + plot(R+A*(1+cos(theta))/2, (theta,0,2*pi),color="red")
```

```python
s = (r*exp(I*theta) + x)/2
```

```python
s1 = s.subs(solutions[0]).subs(a = A, r = R)
```

```python
s2 = s.subs(solutions[1]).subs(a = A, r = R)
```

```python
show(s1)
```

```python
P = R + A/2
Q = R/2
parametric_plot((s1.real_part(),s1.imag_part()),(theta,0,2*pi)) + \
implicit_plot(((x+P/2)*2/P)^2 + (y/Q)^2 == 1, (x,-P,0), (y, -Q, Q), color = "red")
```

```python
show(s2)
```

```python
P = R + A/2
Q = R/2
parametric_plot((s2.real_part(),s2.imag_part()),(theta,0,2*pi)) + \
implicit_plot(((x-P/2)*2/P)^2 + (y/Q)^2 == 1, (x,0,P), (y, -Q, Q), color = "red")
```
